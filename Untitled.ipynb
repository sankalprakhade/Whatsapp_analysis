```python
def find_user(s):
    s=str(s)
    try:
        x=s.index(":")
        return s[:x]
    except:
        return 'f'
    
whatsapp['User']=list(map(find_user,whatsapp['msg'].values))
```

```python
def find_msg(s):
    s=str(s)
    try:
        x=s.index(":")
    except:
        x=0
    return s[x+1:]
whatsapp['msg']=list(map(find_msg,whatsapp['msg'].values))
```

# EDA

Which user has most messages?

```python
c = whatsapp.groupby('User').count()
c.plot.pie(y = 'Message', figsize=(10,8),legend=False,ylabel='',title='Message Count',explode = (0, 0, 0.1, 0, 0, 0,0,0.1),
           shadow = True,autopct='%1.0f%%');
```

## Most Active hours

```python
whatsapp['hours'] = whatsapp['Date/Time'].apply(lambda x: x.hour)
```

```python
z=whatsapp.groupby('hours').count()['Message']
sns.set_style('darkgrid')
sns.set_palette('Dark2')
color=['cornflowerblue' if i==3 else 'lightblue' for i in range(24)]
plt.figure(figsize=(9,6))
plt.bar(z.index,z, color=color)
```

## Day/Month wise

```python
days = ["Monday", "Tuesday", "Wednesday",
        "Thursday", "Friday", "Saturday", "Sunday"]
whatsapp['Day'] = whatsapp['Date/Time'].apply(lambda x: days[x.weekday()])
whatsapp.head()
```

```python
c=whatsapp.groupby('Day')['Message'].count()
c.plot.pie(y = 'Message', figsize=(10,8),legend=False,ylabel='',title='Number of Messages',explode = (0.1, 0, 0, 0, 0, 0,0.1),
           shadow = True,autopct='%1.0f%%')
```

```python
months = ["Jan", "Feb", "Mar","Apr","May",
        "Jun", "Jul", "Aug", "Sept",'Oct','Nov','Dec']
whatsapp['Month'] = whatsapp['Date/Time'].apply(lambda x: months[x.month-1])
whatsapp['Year'] = whatsapp['Date/Time'].apply(lambda x: x.year)
whatsapp.head()
```

```python
d=whatsapp.groupby(['Year','Month']).count()
```

```python
d.plot(y='Message',figsize=(8,6));
```
